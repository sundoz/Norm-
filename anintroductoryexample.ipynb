```python
from numpy.random import uniform

def binomial_rv(n, p):
    c = 0
    for i in range(n):
        U = uniform()
        if U < p:
            c+=1
    return c

binomial_rv(10, 0.5)
```

```python
#4
import numpy as np
from numpy.random import uniform
n=10000
z=0
for i in range(n):
    x,y=uniform(),uniform()
    d=np.sqrt((x-0.5)**2+(y-0.5)**2)
    if d<0.5:
        z+=1
z=z/n
print(z*4)
```
